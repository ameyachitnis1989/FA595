```python
# Import libraries for the code
import requests 
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
import nltk
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
Name=[]                                         #Create list for Name
Purpose=[]                                      #Create list for purpose  
pattern1 = re.compile('Name:')
pattern2 = re.compile('Purpose:')

#Run loop 50 times
for i in range(50):
    source= requests.get("http://3.85.131.173:8000/random_company")
    com_html= BeautifulSoup(source.text,'html.parser')    
    Name.append(com_html.find('li',text=pattern1).text.replace("Name: ",""))
    Purpose.append(com_html.find('li',text=pattern2).text.replace("Purpose: ",""))
    
#Write Company details in df
com_df = pd.DataFrame()
com_df['Name']  = Name
com_df['Purpose']  = Purpose

#Write text file for team to use
np.savetxt(r'E:/FA595/Assign1.txt',com_df.values,fmt='%s',delimiter=';',header="Name;Purpose")

```

```python
#Read files received from team including the initial file created
senti_df = pd.read_csv(
    'E:/FA595/Assign1.txt', sep=";")

senti_df=senti_df.append(pd.read_excel('E:/FA595/Rishi.xlsx').rename({'Company Name': '# Name'}, axis='columns')) 
senti_df=senti_df.append(pd.read_csv('E:/FA595/Yash.csv').rename({'Name': '# Name'}, axis='columns')) 
senti_df=senti_df.append(pd.read_csv('E:/FA595/Jugal.csv').rename({'Name': '# Name'}, axis='columns')) 

```

```python
#Sentiment Analysis using VADER
analyzer= SentimentIntensityAnalyzer()

#Get all compunents of the sentiment for each row
senti_df['compound'] = [analyzer.polarity_scores(x)['compound'] for x in senti_df['Purpose']]
senti_df['neg'] = [analyzer.polarity_scores(x)['neg'] for x in senti_df['Purpose']]
senti_df['neu'] = [analyzer.polarity_scores(x)['neu'] for x in senti_df['Purpose']]
senti_df['pos'] = [analyzer.polarity_scores(x)['pos'] for x in senti_df['Purpose']]

#Sort according to sentiment of the putpose
senti_df=senti_df.sort_values('compound',ascending=False)
```
